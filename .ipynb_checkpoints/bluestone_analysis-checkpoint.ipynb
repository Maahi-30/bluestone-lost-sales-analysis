```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')  # Just makes seaborn plots look cleaner
```

```python
df = pd.read_excel('data/LOS_Base_Data.xlsx', nrows=300)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Submission Date UTC +05:30 (dd/mm/yyyy)', 'Month'])
```

```python
df['Remarks'] = df['Remarks'].fillna('')
```

```python
df['Store'] = df['Store'].str.strip().str.title()
df['ItemCategory'] = df['ItemCategory'].str.strip().str.title()
df['DesignCategory'] = df['DesignCategory'].str.strip().str.title()
```

```python
df.rename(columns={
    'Store_Location_Name': 'store',
    'Product_Category': 'category',
    'Net_Sales': 'sales'
}, inplace=True)
```

```python
df
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('[^A-Za-z0-9_]+', '', regex=True)
```

```python
df.head(5)
```

```python
df.rename(columns={'date_ddmmyyyy': 'date'}, inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip().str.lower()
```

```python
# Checking for fully duplicated rows (i.e., rows with exactly same values across all columns)
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['other_subcategory'], inplace=True)
```

```python
df.drop(columns=['other_categories'], inplace=True)
```

```python
df.drop(columns=['remarks'], inplace=True)
```

```python
df.head(5)
```

```python
# Group by category & count number of records
category_counts = df.groupby(['itemcategory', 'designcategory', 'design_sub_category']).size().reset_index(name='count')

# Sort to find top combinations
top_categories = category_counts.sort_values(by='count', ascending=False)

# Display top 10
top_categories.head(10)
```

```python
store_item_df = df.groupby(['store', 'itemcategory']).size().reset_index(name='count')
store_item_df.sort_values('count', ascending=False).head(10)
```

```python
store_price_df = df.groupby(['store', 'frozenpricerangebroad']).size().reset_index(name='count')
store_price_df.sort_values('count', ascending=False).head(10)
```

```python
store_total_df = df['store'].value_counts().reset_index()
store_total_df.columns = ['store', 'lost_sales_count']
store_total_df.head(10)
```

```python
top_5_stores = store_total_df.head(5)
bottom_3_stores = store_total_df.tail(3)
```

```python
top_5_stores
```

```python
bottom_3_stores
```

```python
# Define your top and bottom stores
top_stores = ['wave one, noida', 'omr thoraipakkam, chennai', 'east coast road, chennai', 'nexus, hyderabad', 'oberoi mall, mumbai']
bottom_stores = ['centrio mall, dehradun', 'viviana mall, thane', 'pacific mall, new delhi']

# Filter the main DataFrame
top_df = df[df['store'].isin(top_stores)]
bottom_df = df[df['store'].isin(bottom_stores)]
```

```python
top_demand = top_df.groupby(['itemcategory', 'designcategory', 'design_sub_category']).size().reset_index(name='count')
top_demand = top_demand.sort_values(by='count', ascending=False).head(10)
top_demand
```

```python
bottom_demand = bottom_df.groupby(['itemcategory', 'designcategory', 'design_sub_category']).size().reset_index(name='count')
bottom_demand = bottom_demand.sort_values(by='count', ascending=False).head(10)
bottom_demand
```

```python
# Convert to sets of tuples for easier comparison
top_set = set(tuple(x) for x in top_demand[['itemcategory', 'designcategory', 'design_sub_category']].values)
bottom_set = set(tuple(x) for x in bottom_demand[['itemcategory', 'designcategory', 'design_sub_category']].values)

# Find popular combinations in top stores that are missing in bottom stores
missing_in_bottom = top_set - bottom_set
missing_in_bottom
```

```python
# Assume top_demand is the DataFrame you've already created
plt.figure(figsize=(12, 6))
sns.barplot(data=top_demand, x='count', y='design_sub_category', hue='itemcategory')
plt.title('Top 10 High-Demand Jewellery Designs')
plt.xlabel('Number of Lost Sales')
plt.ylabel('Design Sub Category')
plt.legend(title='Item Category')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=top_5_stores, x='lost_sales_count', y='store', hue='store', palette='crest', legend=False)
plt.title('Top 5 Stores by Lost Sales')
plt.xlabel('Lost Sales Count')
plt.ylabel('Store')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(
    data=bottom_3_stores,
    x='lost_sales_count',
    y='store',
    hue='store',
    legend=False,
    palette='flare'
)

plt.title('Bottom 3 Stores by Lost Sales')
plt.xlabel('Lost Sales Count')
plt.ylabel('Store')
plt.tight_layout()
plt.show()
```

```python
top_demand['design_combo'] = (
    top_demand['itemcategory'] + ' - ' +
    top_demand['designcategory'] + ' - ' +
    top_demand['design_sub_category']
)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_demand, y='design_combo', x='count', palette='crest')
plt.title('Top 10 Lost Sales Design Combinations')
plt.xlabel('Lost Sales Count')
plt.ylabel('Design Combo')
plt.tight_layout()
plt.show()
```

```python
bottom_demand['design_combo'] = (
    bottom_demand['itemcategory'] + ' - ' +
    bottom_demand['designcategory'] + ' - ' +
    bottom_demand['design_sub_category']
)

plt.figure(figsize=(8, 4))
sns.barplot(data=bottom_demand, y='design_combo', x='count', palette='flare')
plt.title('Bottom 3 Lost Sales Design Combinations')
plt.xlabel('Lost Sales Count')
plt.ylabel('Design Combo')
plt.tight_layout()
plt.show()
```

```python
top_demand['tag'] = 'Top Demand'
bottom_demand['tag'] = 'Bottom Demand'

full_demand = pd.concat([top_demand, bottom_demand])

plt.figure(figsize=(10, 8))
sns.barplot(data=full_demand, y='design_combo', x='count', hue='tag', dodge=False,
            palette={'Top Demand': 'green', 'Bottom Demand': 'red'})
plt.title('High vs Low Demand Design Combinations')
plt.xlabel('Lost Sales Count')
plt.ylabel('Design Combo')
plt.legend(title='Demand Type')
plt.tight_layout()
plt.show()
```
